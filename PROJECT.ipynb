## TRAINING REGRESSION MODEL ##

```python
#training Logistic Regression Model
# Map Outcome to 'Low Level' and 'High Level'
data['Outcome'] = data['Outcome'].map({0: 'Low Level', 1: 'High Level'})

# Select features and target
X = data[['BloodPressure', 'SkinThickness', 'DiabetesPedigreeFunction']]
y = data['Outcome'].map({'Low Level': 0, 'High Level': 1})  # Convert back to 0 and 1 for model training

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**initialising the model**

```python
# Initialize and train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Calculate accuracy on the test set
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

```

## PREDICTION BASED BY MODEL ##

```python
# Function to predict diabetes level based on user input
def predict_diabetes(bp, skin_thickness, pedigree_function):
    prediction = model.predict([[bp, skin_thickness, pedigree_function]])
    return "High Level" if prediction == 1 else "Low Level"

# Example usage: Manually enter the values
bp_input = float(input("Enter Blood Pressure: "))
skin_thickness_input = float(input("Enter Skin Thickness: "))
pedigree_function_input = float(input("Enter Diabetes Pedigree Function: "))

# Get prediction
result = predict_diabetes(bp_input, skin_thickness_input, pedigree_function_input)
print(f"Based on the input values, the model predicts: {result}")
#print the accuracy of the model
print(f'Accuracy of the model: {accuracy:.2f}')
```
