# reading csv file 


```python
from matplotlib import pyplot as plt
sm=pd.read_csv('student_marks.csv')
sm
```

```python
sm.head(2)
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
sm
```

```python
sm['Percentage']=sm['Total']/5
sm
```

```python
#Grading
# A=100-80 
# B=80-60
# C=60-50
# D=50-40
# E=40-30
# F=0-30
def calc_grade(p):
    if p<=100 and p>=80:
        return 'A'
    elif p<80 and p>=60:
        return 'B'
    elif p<60 and p>=50:
        return 'C'
    elif p<50 and p>=40:
        return 'D'
    elif p<40 and p>=30:
        return 'E'
    elif p<30 and p>=0:
        return 'F'
    else:
        return 'X'

sm['Grade']=sm['Percentage'].apply(calc_grade) #apply for applying somefunction on some attribute
sm
    
    
    
```

```python
temp=sm.head()
temp
```

```python
 #1 xaxis and 2nd yaxis
#for changing color we have to make array and allot different colours

for sub in ['English','Hindi','Maths','Science','SST']:
    plt.figure(figsize=(10,6))
    plt.plot(temp['Name'],temp[sub],marker='o',color='red',label=sub) #marker brings o in the graph
    plt.title("Marks in Different Subjects")
    plt.xlabel('Name')
    plt.ylabel('Marks')
    plt.legend()
    plt.grid(True)
    plt.show()
```
